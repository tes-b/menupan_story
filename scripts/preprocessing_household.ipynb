# 3차 전처리

```python
df = pd.read_csv("../data/지역별_세대원수별_세대수_1.csv",index_col=0)
# print(df_hh.shape)
df.head()
```

```python
def split_addr(x):
    addr = x.split()
    # size = len(addr)
    sido, sigu, sigu2, dongmyun = "", "", "", ""
    for area in addr:
        if dongmyun != "": break

        lw = area[-1]
        if lw == "도":
            if sido == "":
                sido = area
        elif lw == "시":
            if sido == "":
                sido = area
            elif sigu == "":
                sigu = area
            elif sigu2 == "":
                sigu2 = area
        elif lw == "구":
            if sigu == "":
                sigu = area
            elif sigu2 == "":
                sigu2 = area
        elif lw in ["동","읍","면"]:
            if dongmyun == "":
                dongmyun = area
    
    return sido, sigu, sigu2, dongmyun    
```

```python
df["sido"],df["sigu"],df["sigu2"],df["dongmyun"] = zip(*df["addr"].map(split_addr))
df.head()
```
