```python
import pandas as pd
```

```python

data = {"code":[],
            "sido":[],
            "sigu":[],
            "sigu2":[],
            "dongmyun":[]
         }

with open("../data/법정동코드 전체자료.txt","r",encoding='cp949') as file:
    codes = file.readlines()
    
    for i,line in enumerate(codes):
        if i == 0 : continue
        
        addr = line.split()

        if addr[-1] == "폐지": continue
        code = int(addr[0][:-2])
        addr = addr[1:]

        sido, sigu, sigu2, dongmyun = "", "", "", ""

        for area in addr:
            if dongmyun != "": break

            lw = area[-1]
            if lw == "도":
                if sido == "":
                    sido = area
            elif lw == "시":
                if sido == "":
                    sido = area
                elif sigu == "":
                    sigu = area
                elif sigu2 == "":
                    sigu2 = area
            elif lw == "구":
                if sigu == "":
                    sigu = area
                elif sigu2 == "":
                    sigu2 = area
            elif lw in ["동","읍","면"]:
                if dongmyun == "":
                    dongmyun = area
        
        # if dongmyun == "" : continue

        data["code"].append(code)
        data["sido"].append(sido)
        data["sigu"].append(sigu)
        data["sigu2"].append(sigu2)
        data["dongmyun"].append(dongmyun)

        # new_row = {
        #     "code":code,
        #     "sido":sido,
        #     "sigu":sigu,
        #     "sigu2":sigu2,
        #     "dongmyun":dongmyun
        #  }

        # df.iloc[i] = new_row
        # df.append(new_row,ignore_index=True)

df = pd.DataFrame(data,columns=["code","sido","sigu","sigu2","dongmyun"])
df.head()

# df.head()
```

```python
df.to_csv("../data/법정동코드_전체자료.csv")
```
